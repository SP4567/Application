```python

# Load CSV data into the products table in SQLite
import pandas as pd
import sqlite3

# Path to your CSV file
csv_file = 'C:\\Users\Suyash Pandey\Desktop\Application\E-commerce_App\src\products.csv' 
df = pd.read_csv(csv_file, encoding='utf-8')

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect('ecommerce.db')
cursor = conn.cursor()

# Create the products table if it doesn't exist
cursor.execute('''
CREATE TABLE IF NOT EXISTS products (
    id INTEGER PRIMARY KEY,
    cost REAL,
    category TEXT,
    name TEXT,
    brand TEXT,
    retail_price REAL,
    department TEXT,
    sku TEXT UNIQUE,
    distribution_center_id INTEGER
)
''')
conn.commit()

# Load data into the products table
df.to_sql('products', conn, if_exists='replace', index=False)

# Preview the first 5 rows to verify
result = pd.read_sql('SELECT * FROM products LIMIT 5', conn)
print(result)

# Close the connection
conn.close()
```
